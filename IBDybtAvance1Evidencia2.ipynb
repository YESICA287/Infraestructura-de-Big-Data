```python

import pyspark

# Create a SparkContext in local mode
sc = pyspark.SparkContext("local")

# Create a SqlContext from the SparkContext
sqlContext = pyspark.SQLContext(sc)
```

```python
data = [
    (1, 'Carolina', 'Romero', 20, 90),
    (2, 'Emiliano', 'Flores', 21, 87),
    (3, 'Miguel', 'Esteves', 20, 89),
    (4, 'Gerardo', 'Garcia', 21, 79),
    (5, 'Benjamin', 'Hernandez', 21, 84),
    (6, 'Gerardo', 'Chavez', 22, 78),
    (7, 'Kathya', 'Gomez', 22, 82),
    (8, 'Asdru', 'Diaz', 20, 91),
    (9, 'Guillermo', 'Lopez', 20, 84),
    (10, 'Gustavo', 'Mathieu', 22, 81),
    (11, 'Jorge', 'Estrada', 22, 70),
    (12, 'Yesica', 'Bautista', 29, 50),
]

# Convert a local data set into a DataFrame
df = sqlContext.createDataFrame(data, ['ID', 'Nombre', 'Apellido', 'Edad', 'Calificacion'])

# Convert to a Pandas DataFrame for easy display
df.toPandas()
```

```python
df.createOrReplaceTempView("mytable")
```

```python
# Peform a simple select from the table
results = sqlContext.sql("select * from mytable")

# Convert the results to a Pandas DataFrame for easy viewing
results.toPandas()
```

```python
# Muestra la estructura del DataFrame
df.printSchema()
# Perform a query with a where clause and orde
results = sqlContext.sql("SELECT * FROM mytable WHERE ID <= 13 ORDER BY ID DESC")

# Convert the results to a Pandas DataFrame for easy viewing
results.toPandas()
```

```python
# Perform a more complex query on the table
results = sqlContext.sql("select Edad, count(*) as count, avg(Calificacion) as avg, sum(Edad) as sum from mytable group by Edad")

# Convert the results to a Pandas DataFrame for easy viewing
results.toPandas()
```
